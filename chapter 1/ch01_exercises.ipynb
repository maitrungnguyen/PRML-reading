\begin{align*}
p_y(y) &= p_x(x) \left| \frac{d x}{d y} \right| \\
&= p_x(g(y)) \left| g'(y) \right|
\end{align*}

To find the maximum of the density function $p_y(y)$ we set its derivative to zero:
\begin{align*}
\frac{d p_y(y)}{d y} &= \frac{d}{d y} \left[ p_x(g(y)) \left| g'(y) \right| \right] = 0
\end{align*}

Using the product rule we have:
\begin{align*}
\frac{d p_y(y)}{d y} &= \left| g'(y) \right| \frac{d}{d y} p_x(g(y)) + p_x(g(y)) \frac{d}{d y} \left| g'(y) \right| = 0
\end{align*}

Using the chain rule on the first term:
\begin{align*}
\frac{d p_y(y)}{d y} &= \left| g'(y) \right| p_x'(g(y)) g'(y) + p_x(g(y)) \frac{d}{d y} \left| g'(y) \right| = 0
\end{align*}

We assume that $g(y)$ is monotone increasing (monotone decreasing is the same), so that $g'(y) > 0$ and $\left| g'(y) \right| = g'(y)$:
\begin{align*}
\frac{d p_y(y)}{d y} &= g'(y) p_x'(g(y)) g'(y) + p_x(g(y)) \frac{d}{d y} g'(y) = 0 \\
&= (g'(y))^2 p_x'(g(y)) + p_x(g(y)) g''(y) = 0
\end{align*}

Rearranging gives:
\begin{align*}
(g'(y))^2 p_x'(g(y)) &= - p_x(g(y)) g''(y)
\end{align*}

With $\hat{x}$ is the maximum of $p_x(x)$, we have $p_x'(\hat{x}) = 0$. Thus if we set $g(\hat{y}) = \hat{x}$ then the left hand side is zero. For the right hand side to also be zero we need $g''(\hat{y}) = 0$. This is not true in general, so in general the maximum of $p_y(y)$ is not at $g^{-1}(\hat{x})$. 

However, if $g(y)$ is linear then $g''(y) = 0$ for all $y$. Thus if $g(y)$ is linear then the maximum of $p_y(y)$ is at $g^{-1}(\hat{x})$.




We want to prove with $x$ and $z$ independent, then:
\begin{align*}
E[x+z] &= E[x] + E[z] \\
Var[x+z] &= Var[x] + Var[z]
\end{align*}

Recall the definitions of expectation and variance:
\begin{align*}
E[x] &= \int x p_x(x) d x \\
Var[x] &= \int (x - E[x])^2 p_x(x) d x
\end{align*}
Since $x$ and $z$ are independent, we have:
\begin{align*}
p_{x,z}(x,z) &= p_x(x) p_z(z)
\end{align*}

To prove the expectation result:
\begin{align*}
E[x+z] &= \int \int (x + z) p_{x,z}(x,z) d x d z \\
&= \int \int (x + z) p_x(x) p_z(z) d x d z \\
&= \int \left[ \int x p_x(x) d x \right] p_z(z) d z + \int \left[ \int z p_z(z) d z \right] p_x(x) d x \\
&= E[x] \int p_z(z) d z + E[z] \int p_x(x) d x \\
&= E[x] + E[z]
\end{align*}

To prove the variance result:
\begin{align*}
Var[x+z] &= \int \int \left( (x + z) - E[x+z] \right)^2 p_{x,z}(x,z) d x d z \\
&= \int \int \left( (x + z) - (E[x] + E[z]) \right)^2 p_x(x) p_z(z) d x d z \\
&= \int \int \left( (x - E[x]) + (z - E[z]) \right)^2 p_x(x) p_z(z) d x d z \\
&= \int \int \left( (x - E[x])^2 + (z - E[z])^2 + 2(x - E[x])(z - E[z]) \right) p_x(x) p_z(z) d x d z \\
&= \int \left[ \int (x - E[x])^2 p_x(x) d x \right] p_z(z) d z + \int \left[ \int (z - E[z])^2 p_z(z) d z \right] p_x(x) d x \\
&\quad + 2 \int \left[ \int (x - E[x]) p_x(x) d x \right] (z - E[z]) p_z(z) d z \\
&= Var[x] \int p_z(z) d z + Var[z] \int p_x(x) d x + 2 \cdot 0 \\
&= Var[x] + Var[z]
\end{align*}

\end{align*}

\begin{align*}

\end{align*}

\begin{align*}

Given a loss matrix with elements $L_{kj}$ representing the loss incurred for classifying a pattern as class $C_k$ when it actually belongs to class $C_j$, we want to show that the minimum expected loss is achieved by choosing the class $C_k$ that minimizes the loss:
\begin{align*}
\sum_{k} L_{kj} p(C_k | \mathbf{x})
\end{align*}

When the loss matrix is given as $L_{kj} = 1 - I_{kj}$ where $I_{kj}$ is the element of the identity matrix, then the expected loss becomes:
\begin{align*}
\sum_{k} L_{kj} p(C_k | \mathbf{x}) &= \sum_{k} (1 - I_{kj}) p(C_k | \mathbf{x}) \\
&= \sum_{k} p(C_k | \mathbf{x}) - \sum_{k} I_{kj} p(C_k | \mathbf{x}) \\
&= 1 - p(C_j | \mathbf{x})
\end{align*}

Thus minimizing the expected loss is equivalent to maximizing the posterior probability $p(C_j | \mathbf{x})$. The intuitive interpretation is that we want to choose the class that is most probable given the observed data $\mathbf{x}$ in order to minimize the expected loss.


Consider an M-state discrete random variable $x$, use Jensen's inequality to show that the entropy $H[x]$ satisfies:
\begin{align*}
H[x] &\leq \ln M
\end{align*}
We recall the definition of entropy for a discrete random variable:
\begin{align*}
H[x] &= - \sum_{i=1}^M p(x_i) \ln p(x_i)
\end{align*}

Consider $f(x) = -x \ln x$ has second derivative:
\begin{align*}
f'(x) &= - \ln x - 1 \\
f''(x) &= - \frac{1}{x}
\end{align*}
Since $f''(x) < 0$ for $x > 0$, $f(x)$ is concave. By Jensen's inequality for a concave function:

\begin{align*}
f\left( \frac{\sum_{i} p(x_i)}{M} \right) &\geq \frac{1}{M} \sum_{i} f(p(x_i)) \\
f\left( \frac{1}{M} \right) &\geq \frac{1}{M} \sum_{i} -p(x_i) \ln p(x_i) \\
- \frac{1}{M} \ln \left( \frac{1}{M} \right) &\geq \frac{1}{M} H[x] \\
\ln M &\geq H[x]
\end{align*}


Consider two variables $x$ and $y$. We want to show that the differential entropy satisfies:
\begin{align*}
H[x,y] &\leq H[x] + H[y]
\end{align*}
Recall the definition of differential entropy for continuous random variables:
\begin{align*}
H[x, y] &=  H[x] + H[y|x] \\
&\leq H[x] + H[y]
\end{align*}
The inequality follows from the fact that conditioning reduces entropy: $H[y|x] \leq H[y]$. Thus we have shown that the joint differential entropy is less than or equal to the sum of the individual differential entropies.

The equality holds when $H[y|x] = H[y]$, which occurs when $x$ and $y$ are independent random variables. In that case, knowing $x$ provides no information about $y$, so the conditional entropy $H[y|x]$ equals the marginal entropy $H[y]$.

This is the form of a Gaussian distribution. To determine the Lagrange multipliers $\lambda_1$, $\lambda_2$, and $\lambda_3$, we apply the constraints:
1. Normalization: $\int p(x) d x = 1$
2. Mean: $\int p(x) x d x = \mu$
3. Variance: $\int p(x) (x - \mu)^2 d x = \sigma^2$

Integrating these constraints with the expression for $p(x)$ leads to a system of equations for the Lagrange multipliers:
\begin{align*}
\int \exp \left( \lambda_1 - 1 + \lambda_2 x + \lambda_3 (x - \mu)^2 \right) d x &= 1 \\
\int x \exp \left( \lambda_1 - 1 + \lambda_2 x + \lambda_3 (x - \mu)^2 \right) d x &= \mu \\
\int (x - \mu)^2 \exp \left( \lambda_1 - 1 + \lambda_2 x + \lambda_3 (x - \mu)^2 \right) d x &= \sigma^2
\end{align*}

We recognize that these integrals correspond to the moments of a Gaussian distribution. Solving these constraints leads to the values of the Lagrange multipliers:
\begin{align*}
\lambda_3 &= -\frac{1}{2 \sigma^2} \\
\lambda_2 &= 0 \\
\lambda_1 &= 1 - \frac{1}{2} \ln(2 \pi \sigma^2)
\end{align*}    
Substituting these back into the expression for $p(x)$ gives the Gaussian distribution:
\begin{align*}
p(x) &= \frac{1}{\sqrt{2 \pi \sigma^2}} \exp \left( -\frac{(x - \mu)^2}{2 \sigma^2} \right)
\end{align*}

We have shown that the probability density function that maximizes the entropy subject to the given constraints is the Gaussian distribution with mean $\mu$ and variance $\sigma^2$.


To prove the convex function inequality, by induction on $n$:
For $n = 2$, the inequality holds by the definition of convexity:
\begin{align*}
f(\lambda x_1 + (1-\lambda) x_2) &\le \lambda f(x_1) + (1-\lambda) f(x_2)
\end{align*}

Assume the inequality holds for $n = k$:
\begin{align*}
f\left( \sum_{i=1}^k \lambda_i x_i \right) &\le \sum_{i=1}^k \lambda_i f(x_i)
\end{align*}

We want to show it holds for $n = k + 1$:
\begin{align*}  
f\left( \sum_{i=1}^{k+1} \lambda_i x_i \right) &= f\left( \left( \sum_{i=1}^k \lambda_i \right) \cdot \frac{\sum_{i=1}^k \lambda_i x_i}{\sum_{i=1}^k \lambda_i} + \lambda_{k+1} x_{k+1} \right) \\
&\le \left( \sum_{i=1}^k \lambda_i \right) f\left( \frac{\sum_{i=1}^k \lambda_i x_i}{\sum_{i=1}^k \lambda_i} \right) + \lambda_{k+1} f(x_{k+1}) \\
&\le \left( \sum_{i=1}^k \lambda_i \right) \cdot \frac{\sum_{i=1}^k \lambda_i f(x_i)}{\sum_{i=1}^k \lambda_i} + \lambda_{k+1} f(x_{k+1}) \\
&= \sum_{i=1}^{k+1} \lambda_i f(x_i)
\end{align*}

Thus by induction, the convex function inequality holds for all $n \ge 2$.

First, we recall the definition of $I(\mathbf{x}, \mathbf{y})$:
\begin{align*}
I(\mathbf{x}, \mathbf{y}) &= - \int \int p(\mathbf{x}, \mathbf{y}) \ln \left( \frac{p(\mathbf{x})p(\mathbf{y})}{p(\mathbf{x}, \mathbf{y})} \right) d \mathbf{x} d \mathbf{y}
\end{align*}

We want to show that $I(\mathbf{x}, \mathbf{y}) = H[\mathbf{x}] - H[\mathbf{x}|\mathbf{y}]$. We start by expanding the logarithm in the definition of mutual information:
\begin{align*}
I(\mathbf{x}, \mathbf{y}) &= - \int \int p(\mathbf{x}, \mathbf{y}) \left[ \ln p(\mathbf{x}) + \ln p(\mathbf{y}) - \ln p(\mathbf{x}, \mathbf{y}) \right] d \mathbf{x} d \mathbf{y} \\
&= - \int \int p(\mathbf{x}, \mathbf{y}) \ln p(\mathbf{x}) d \mathbf{x} d \mathbf{y} - \int \int p(\mathbf{x}, \mathbf{y}) \ln p(\mathbf{y}) d \mathbf{x} d \mathbf{y} + \int \int p(\mathbf{x}, \mathbf{y}) \ln p(\mathbf{x}, \mathbf{y}) d \mathbf{x} d \mathbf{y}
\end{align*}

The first term is the negative joint entropy:

\begin{align*}
- \int \int p(\mathbf{x}, \mathbf{y}) \ln p(\mathbf{x}, \mathbf{y}) d \mathbf{x} d \mathbf{y} &= H[\mathbf{x}, \mathbf{y}]
\end{align*}

The second term simplifies to the marginal entropy of $\mathbf{x}$:

\begin{align*}
- \int \int p(\mathbf{x}, \mathbf{y}) \ln p(\mathbf{x}) d \mathbf{x} d \mathbf{y} &= - \int p(\mathbf{x}) \ln p(\mathbf{x}) d \mathbf{x} = H[\mathbf{x}]
\end{align*}

The third term simplifies to the marginal entropy of $\mathbf{y}$:

\begin{align*}
- \int \int p(\mathbf{x}, \mathbf{y}) \ln p(\mathbf{y}) d \mathbf{x} d \mathbf{y} &= - \int p(\mathbf{y}) \ln p(\mathbf{y}) d \mathbf{y} = H[\mathbf{y}]
\end{align*}

Putting it all together, we have:
\begin{align*}
I(\mathbf{x}, \mathbf{y}) &= - H[\mathbf{x}, \mathbf{y}] + H[\mathbf{x}] + H[\mathbf{y}]
\end{align*}

Using the chain rule for entropy, we know that:
\begin{align*}
H[\mathbf{x}, \mathbf{y}] &= H[\mathbf{x}] + H[\mathbf{y}|\mathbf{x}]
\end{align*}

Substituting this into the expression for mutual information gives:
\begin{align*}
I(\mathbf{x}, \mathbf{y}) &= - (H[\mathbf{x}] + H[\mathbf{y}|\mathbf{x}]) + H[\mathbf{x}] + H[\mathbf{y}] \\
&= H[\mathbf{y}] - H[\mathbf{y}|\mathbf{x}]
\end{align*}

Similarly, we can also express mutual information as:
\begin{align*}
I(\mathbf{x}, \mathbf{y}) &= H[\mathbf{x}] - H[\mathbf{x}|\mathbf{y}]
\end{align*}
